### (b) Keep datasets 1 and 2 in folders bending1 and bending 2, as well as datasets 1, 2, and 3 in other folders as test data and other datasets as train data.


```python
def listdir_nohidden(path):
 return sorted(glob.glob(os.path.join(path, '*')))

# dictionary[folderName] = [dataset1, sataset2]
# store dataset1 as dataFrame
dirPath = 'AReM/'
folderNameList = ['bending1', 'bending2', 'cycling', 'lying', 'sitting', 'standing', 'walking']
columnNameList = ['time', 'avg_rss12', 'var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23']
# Store all data with index: folder name
datasetDict = dict()
setNum = 2
count = 0
trainSet = list()
testSet = list()
for folder in folderNameList:
    datasetList=[]
    path = dirPath + folder
    # Filter hidden file
    arry = listdir_nohidden(path)
    # Sorted file by number
    arry = sorted(arry, key=lambda x: int(re.match('\D*(\d+)', x).group(1)))
    
    if folder not in ['bending1', 'bending2']:
        setNum = 3
    count = 0
    for file in arry:
        data = pd.read_csv(file, skiprows=5, header=None, names=columnNameList)
        data = data.replace(np.nan, 0)
        datasetList.append(data)
        count += 1
        if count <= setNum:
#             print('testdata', folder, file)
            testSet.append(data)
        else:
#             print('.....traindata', folder, file)
            trainSet.append(data)
    datasetDict[folder] = datasetList

datasetDict
# print(trainSet)
```

#### ii. Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance. You are free to normalize/standardize features or use them directly.

```python
import collections

statCollection = collections.defaultdict(dict)
insNum = 0
for folder in folderNameList:
    for dataset in datasetDict[folder]:
        insNum += 1
        featuresNum = 0
        for col in columnNameList[1:]:
            featuresNum += 1
#             print(featuresNum)
            infor = dataset[col].describe()
            statCollection[insNum]['min' + str(featuresNum)] = infor['min']
            statCollection[insNum]['max' + str(featuresNum)] = infor['max']
            statCollection[insNum]['mean' + str(featuresNum)] = infor['mean']
            statCollection[insNum]['median' + str(featuresNum)] = infor['50%']
            statCollection[insNum]['std' + str(featuresNum)] = infor['std']
            statCollection[insNum]['firstQuartile' + str(featuresNum)] = infor['25%']
            statCollection[insNum]['thirdQuartile' + str(featuresNum)] = infor['75%']
tableInfor = pd.DataFrame(statCollection).T
tableInfor
```

#### iii. Estimate the standard deviation of each of the time-domain features you extracted from the data. Then, use Pythons bootstrapped or any other method to build a 90% bootsrap confidence interval for the standard deviation of each feature.

```python
from astropy.stats import bootstrap
import statistics
from numpy import mean
from numpy import median
from numpy import percentile
import scipy as sp
import scipy.stats

# Calculate 90% bootstrap confidence interval
def bootstrapInterval(dataset, samplingTimes, confidence=0.1): 
    # sampel score by using func
    scores = list()
    # the number of sampling time
    for i in range(samplingTimes):
        indices = np.random.randint(0, len(dataset), size=len(dataset))
        sample = dataset[indices]
        # Using std func to calculate
        sample_result = np.std(sample)
        scores.append(sample_result)
    # Confidence is 90%
    a = 1-confidence
    k1 = int(samplingTimes * a / 2)
    k2 = int(samplingTimes * (1-a/2))
    sampleSorted = sorted(scores)
    lower = sampleSorted[k1]
    higher = sampleSorted[k2]
    return round(lower, 2), round(higher, 2)

def std_confidence_interval(data, confidence=0.9):
    a = np.array(data)
    n = len(a)
    m, se = np.std(a), scipy.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return round(m-h,2), round(m+h, 2)

resultDic = collections.defaultdict(list)
for col in tableInfor:
    # Calculate standard deviation
    resultDic[col].append(statistics.stdev(tableInfor[col]))
    resultDic[col].append(bootstrapInterval(tableInfor[col], 10))
    
print(pd.DataFrame.from_dict(resultDic, orient='index', columns = ['std', 'interval']))
```

#### ii. Break each time series in your training set into two (approximately) equal length time series. Now instead of 6 time series for each of the training instances, you have 12 time series for each training instance. Repeat the experiment in 1(d)i, i.e depict scatter plots of the features extracted from both parts of the time series 1,2, and 12. Do you see any considerable difference in the results with those of 1(d)i?

```python
def breakTimeSeries(dataset, numsOfPart):
    newDataSet = list()
    for df in dataset:
        size = round(len(df)/numsOfPart)
        list_of_dfs = [df.iloc[i:i+size-1,:] for i in range(0, len(df),size)]
        tmpDf = pd.DataFrame()
        for s in range(numsOfPart):
            # rearrange index for concating
            list_of_dfs[s].index = range(len(list_of_dfs[s]))
            # delete 'time' column
            tmpDf = pd.concat([list_of_dfs[s].iloc[:, 1:], tmpDf], axis=1, ignore_index=True)
        newDataSet.append(tmpDf)
#         print(newDataSet)
    return newDataSet

timeSeries = [0, 1, 11]
newTrainSet = breakTimeSeries(trainSet, 2)
drawAllScatterPlot(timeSeries, newTrainSet)
```

#### iii. Break each time series in your training set into l  {1, 2, . . . , 20} time series of approximately equal length and use logistic regression5 to solve the binary classification problem, using time-domain features.

```python
# Recursive Feature Elimination
from sklearn import datasets
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFECV
from sklearn.metrics import f1_score
import warnings
warnings.filterwarnings('ignore')

# there are 6*l columns 
def getTarget(dataset, cntOfBending=9):
    cnt = 0
    trainY = list()
    trainX = list()
    # assume bending is 1, non-bending is 0
    for df in dataset:
    #     print(df)
        cnt += 1
        if cnt <= cntOfBending:
            trainY.append(1)
        else:
            trainY.append(0)
        trainX.append(df)
    return trainX, trainY

def calTimeDomainFeatures(datasetList):
    rowList = list()
    data = list()
    for ds in datasetList:
        for col in ds:
            infor = ds[col].describe()
            rowList.append(infor['mean'])
            rowList.append(infor['50%'])
            rowList.append(infor['std'])
        data.append(rowList)
        rowList = []
    data = pd.DataFrame(data)
#     print(data)
    return data

for l in range(1,21):
    newTrainSet = breakTimeSeries(trainSet, l)
    rawData, trainY = getTarget(newTrainSet)
    trainX = calTimeDomainFeatures(rawData)
    model = LogisticRegression()
    
    # create the RFE model and select best attributes
    rfe = RFECV(model, step = 1, cv = 5, scoring='accuracy')
    rfe = rfe.fit(trainX, trainY)
    F1_Score = f1_score(trainY, rfe.predict(trainX))
    print('L:', l)
    print('F1 train score:', round(F1_Score, 2))
    print('Optimal number of features :', rfe.n_features_)
    print('Best features :', trainX.columns[rfe.support_])
    print('\n')
```
